**NOMBRE DE VICTOIRES À DOMICILE / EXTERIEUR PAR ÉQUIPE**

Les performances des équipes selon le lieu du match sont un critère majeur pour prédire les résultats.

Ce code calcule le nombre de victoires à domicile et à l’extérieur pour chaque club, regroupe les résultats dans un DataFrame unique, puis génère un graphique à barres horizontales.
***Les barres bleues***  représentent ***les victoires à domicile***, tandis que ***les barres orange*** illustrent ***les victoires à l’extérieur.***

```python
# Victoire a domicile
home_wins = match_2013_2022.loc[match_2013_2022['home_club_goals'] > match_2013_2022['away_club_goals']].groupby('home_club_name')['results'].count().reset_index(name='home_wins')
# Victoire a l'exterieur
away_wins = match_2013_2022.loc[match_2013_2022['away_club_goals'] > match_2013_2022['home_club_goals']].groupby('away_club_name')['results'].count().reset_index(name='away_wins')
```

```python
home_wins = home_wins.rename(columns = {'home_club_name':'club_name'})
away_wins = away_wins.rename(columns = {'away_club_name':'club_name'})
```

Fusion des differents dataframes

```python
club_wins = home_wins.merge(away_wins, on = 'club_name' , how = 'outer').fillna(0)
club_wins['total_wins'] = club_wins['home_wins'] + club_wins['away_wins']
```
